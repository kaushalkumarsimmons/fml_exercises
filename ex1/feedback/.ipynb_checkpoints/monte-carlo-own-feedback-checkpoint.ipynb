```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

# 1.1 Data Creation and Visualization

$p(X=x|Y=0)$

$PDF:f_0(x) = 2 - 2x$<br>
$CDF:F_0(x) = 2x - x^2$

#### Compute inverse function

$F_0(F_0^{-1}(u)) = u$<br>
$2F_0^{-1}(u) - (F_0^{-1}(u))^2 = u$<br>
$(F_0^{-1}(u))^2 - 2F_0^{-1}(u) = -u$<br>
$(F_0^{-1}(u))^2 - 2F_0^{-1}(u) + (\frac{2}{2})^2 - (\frac{2}{2})^2 = -u$<br>
$(F_0^{-1}(u) - 1)^2 - 1 = - u$<br>
$(F_0^{-1}(u) - 1)^2 = 1 - u$<br>
$F_0^{-1}(u) - 1 = \pm\sqrt{1-u}$<br>
$F_0^{-1}(u) = 1 \pm\sqrt{1-u}$<br>
$1 + \sqrt{1-u}\not\in[0,1]$<br>
$\Rightarrow F_0^{-1}(u) = 1 - \sqrt{1-u}$

<span style ="color:green;font-weight:bold"> Comment </span >

$p(X=x|Y=0)$

$PDF:f_1(x) = 2x$<br>
$CDF:F_1(x) = x^2$

#### Compute inverse function

$F_1(F_1^{-1}(u)) = u$<br>
$(F_1^{-1}(u))^2 = u$<br>
$F_1^{-1}(u) = \sqrt{u}$

```python
def create_data(N):
    data = {"x":[], "y":[]}
    U = np.random.uniform(0.0, 1.0, N)
    for u in U:
        if random.randint(0, 1):
            data["x"].append(np.sqrt(u))
            data["y"].append(1)
        else:
            data["x"].append(1-np.sqrt(1-u))
            data["y"].append(0)
    
    return data
```

```python
data = create_data(100000)
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15,5), tight_layout=True)
axs[0].hist([x for x, y in zip(data["x"], data["y"]) if y == 0], 10);
axs[0].set_title("Frequency of x for y == 0");
axs[0].set_xlabel("x");
axs[0].set_ylabel("frequency");

axs[1].hist([x for x, y in zip(data["x"], data["y"]) if y == 1], 10);
axs[1].set_title("Frequency of x for y == 1");
axs[1].set_xlabel("x");
axs[1].set_ylabel("frequency");

axs[2].hist(data["y"]);
axs[2].set_title("Distribution of labels");
axs[2].set_xlabel("y");
axs[2].set_ylabel("frequency");
```

# 1.2 Classification by Thresholding

```python
def classifier_A(X, x_t):
    return X > x_t
```

```python
def classifier_B(X, x_t):
    return X < x_t
```

```python
def loss(pred, Y):
    return np.mean(abs(Y - pred))
```

```python
error_rates = {"A":{"0.0":[], "0.2":[], "0.5":[], "0.6":[]}, "B":{"0.0":[], "0.2":[], "0.5":[], "0.6":[]}}

for N in [10, 100, 1000, 10000]:
    for x_t in [0., 0.2, 0.5, 0.6]:
        temp_err_A = []
        temp_err_B = []
        
        for i in range(10):
            data = create_data(N)

            pred_A = classifier_A(np.array(data["x"]), x_t)
            pred_B = classifier_B(np.array(data["x"]), x_t)

            loss_A = loss(pred_A, np.array(data["y"]))
            loss_B = loss(pred_B, np.array(data["y"]))

            temp_err_A.append(loss_A)
            temp_err_B.append(loss_B)
        
        error_rates["A"][str(x_t)].append((np.mean(temp_err_A), np.std(temp_err_A)))
        error_rates["B"][str(x_t)].append((np.mean(temp_err_B), np.std(temp_err_B)))
```

```python
for c in ["A", "B"]:
    labels = ["10", "100", "1000", "10000"]

    x = np.arange(len(labels))
    width = 0.2  

    fig, ax = plt.subplots(figsize=(8,8))
    rects1 = ax.bar(x - 1.5*width, [round(a[0], 4) for a in error_rates[c]["0.0"]], width, 
                    yerr=[round(a[1], 4) for a in error_rates[c]["0.0"]], label="0.0")
    rects2 = ax.bar(x - width/2, [round(a[0], 4) for a in error_rates[c]["0.2"]], width,
                    yerr=[round(a[1], 4) for a in error_rates[c]["0.2"]], label="0.2")
    rects3 = ax.bar(x + width/2, [round(a[0], 4) for a in error_rates[c]["0.5"]], width,
                    yerr=[round(a[1], 4) for a in error_rates[c]["0.5"]], label="0.5")
    rects4 = ax.bar(x + 1.5*width, [round(a[0], 4) for a in error_rates[c]["0.6"]],
                    width, yerr=[round(a[1], 4) for a in error_rates[c]["0.6"]], label="0.6")

    ax.set_ylabel("mean error")
    ax.set_xlabel("dataset size")
    ax.set_title("Error results classifier " + c)
    ax.set_xticks(x)
    ax.set_xticklabels(labels)
    ax.legend()

    ax.bar_label(rects1, padding=-20, rotation="vertical", label_type="center")
    ax.bar_label(rects2, padding=-20, rotation="vertical", label_type="center")
    ax.bar_label(rects3, padding=-20, rotation="vertical", label_type="center")
    ax.bar_label(rects4, padding=-20, rotation="vertical", label_type="center")

    fig.tight_layout()

    plt.show()
```

#### The error standard deviation decreases continously for larger datasets. Hence for the dataset with N=10000 the error standard deviation is the smallest.

# 1.4 Nearest Neighbor Classification

```python
def toy_nearest_neighbour(Xtest, Xtrain, Ytrain):
    dist = []
    label = []
    for x, y in zip(Xtrain, Ytrain):
        dist.append(abs(Xtest - x))
        label.append(y)
    
    return np.array(label)[np.argmin(np.array(dist), axis=0)]
```

```python
def create_data(N):
    X = []
    Y = []
    while True:
        U = np.random.uniform(0.0, 1.0, N)
        for u in U:
            if random.randint(0, 1):
                X.append(np.sqrt(u))
                Y.append(1)
            else:
                X.append(1-np.sqrt(1-u))
                Y.append(0)
                
        s  = np.sum(Y)
        if s != 0 and s != N:
            return X, Y
        else:
            X = []
            Y = []
```

```python
error = {"2":[], "100":[]}

for N in [2, 100]:
    for i in range(100):
        Xtrain, Ytrain = create_data(N)
        Xtest, Ytest = create_data(10000)
        
        pred = toy_nearest_neighbour(Xtest, Xtrain, Ytrain)
        
        error[str(N)].append(loss(pred, Ytest))
```

```python
labels = ["2", "100"]

x = np.arange(len(labels))
width = 0.35  

fig, ax = plt.subplots(figsize=(8,8))
rects1 = ax.bar(0, np.mean(error["2"]), width, label="2")
rects2 = ax.bar(1, np.mean(error["100"]), width, label="100")


ax.set_ylabel("mean error")
ax.set_xlabel("training set size")
ax.set_title("Error results nearest neighbour classifier")
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

ax.bar_label(rects1, padding=-20, rotation='vertical', label_type='center')
ax.bar_label(rects2, padding=-20, rotation='vertical', label_type='center')

fig.tight_layout()

plt.show()
```
